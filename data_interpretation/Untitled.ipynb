```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_binary = open("../data/ALG2_JKB1a_X02a_RADjh1/ALG2_JKB1a_X02a_RADjh1.bxds1.00", "rb")
data = []
for _ in range(10_000_000):
    data.append(int.from_bytes(file_binary.read(2), byteorder='little'))
x = np.array([i for i in range(len(data))])
data = np.array(data)
```

```python
plt.ylabel("Voltage")
plt.scatter(x, data)
plt.ylim([0, np.amax(data)])
plt.show()
```

```python
plt.ylabel("Voltage")
scale = 100
plt.scatter(x[:len(data)//scale], data[:len(data)//scale])
plt.ylim([np.amin(data[:len(data)//scale]), np.amax(data[:len(data)//scale])])
plt.show()
```
